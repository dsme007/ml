```python
df_world = pd.read_csv('awesome-pandas/tutorial/data/world_population_history.csv', sep=',')
df_world.head()
```

```python
drop_cols = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', 'World Development Indicators']
```

```python
df_world = (df_world
    .drop(columns=drop_cols)
    .dropna(axis=1, how='all')
)
```

```python
# UNSTACKING AND STACKING
# THE ABOVE WORLD DATA ISN'T IN THE RIGHT FORMAT
# IT'S OUR JOB TO FORMAT IT USING THE STACK FUNCTION

df_world_tidy = (df_world
                 .columns
#                  .set_index(['Data Source'])
#                  .stack(0)
#                  .rename('Population')
#                  .to_frame()
#                  .reset_index()
#                  .rename(columns={'Data Source': 'Country', 'level_1': 'Year'})
#                  .assign(Year=lambda df:pd.to_datetime(df.Year))
#                  .head()
                )

df_world_tidy


```

```python
# UNSTACKING

# slice(None) = ignore the first column level (i.e, the population column)
to_plot = (df_world_tidy
          .set_index(['Country', 'Year'])
          .unstack(level=0)
          .loc[:, (slice(None), ['Norway', 'Sweden'])]
          )

to_plot
```

```python
# PIVOTING AND MELTING
# these are more versatile versions of stacking and unstacking

df_world_pivot = (df_world_tidy
                 .pivot(index='Year', columns='Country', values='Population'))

df_world_pivot.iloc[:5, :5]
```

```python
# drop any country with missing values
df_world_pivot = df_world_pivot.dropna(how='all', axis=1)
df_world_pivot.head()
```

```python
# COMPUTE RELATIVE CHANGE SINCE 1960
# THIS IS CALLED BROADCASTING IN NUMPY
first_row = df_world_pivot.iloc[0, :]
df_world_rel = (df_world_pivot/first_row)
df_world_rel
```

```python
# WHEN PIVOTED, COMPUTING THE RELATIVE CHANGE IS EASY!!!!
# CONVERTING BACK TO TIDY FORMAT CAN BE ACHIEVED USING MELT!
df_world_rel = df_world_rel.reset_index()
df_world_rel
```

```python
df_world_rel = df_world_rel.melt(id_vars = 'Year', value_name='Population')
```

```python
# since we now know the relative change since 1960, we can now use this difference to make a plot
top_n = 5
plot_title = f'Population growth of the {top_n} countries with largest relative change'
countries = (df_world_rel
             .groupby('Country')
             .max()
             .Population
             .nlargest(top_n)
             .index
            )
countries

# plot the population of the above contries
(df_world_rel
 .pivot(index='Year', columns='Country', values='Population')
 .loc[:, countries]
 .plot(figsize=(10, 5), title=plot_title, grid=True, lw=2)
)
```

```python
# THE MELT IS MORE POWERFUL THAN STACK
# UNSTACK AND STACK ARE INVERSE OPERATIONS
# PIVOT AND MELT ARE INVERSE OPERATIONS
```

```python
# MERGING (PANDAS) == JOINING (SQL)
# left join                 right join
# inner join                full join
```

```python
director_means = (df
                 .groupby('director_name')
                 .mean()
                 .round(1)
                 .loc[:, ['imdb_score']]
                 .rename(columns={'imdb_score': 'director_imdb_score'})
                )

df.head()

director_means.head()
```

```python
# LEFT JOIN/MERGE

df.merge(director_means, how='left', left_on='director_name', right_index=True).head()
```

```python
# for every director which has made movies in USA and Canada
# get average imdb scores for both locations

american_directors = (df[df.country.isin(['USA'])]
                      .groupby('director_name')
                      .mean()
                      .imdb_score
                      .to_frame()
                     )

canadian_directors = (df[df.country.isin(['Canada'])]
                      .groupby('director_name')
                      .mean()
                      .imdb_score
                      .to_frame()
                     )

american_directors.merge(canadian_directors, 
                         how='inner', 
                         left_index=True, 
                         right_index=True,
                         suffixes=('_USA', '_Canada')).round(1).head()
```
